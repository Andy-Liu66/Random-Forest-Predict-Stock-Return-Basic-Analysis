## 隨機森林應用於股價報酬之預測 - 基本面分析

* 操作概念：

> 基本分析為眾多股市投資人使用的分析方法之一，亦即透過檢視政治經濟環境、公司財務報表資訊及市場需求調查等等，分析公司未來前景與獲利能力，進而做出買入或賣出股票的選擇，通常此類分析法較多以長期投資的形式進行，較少考量股票市場短期的價格波動，其中基本分析方法最具規律的資訊來源即為公司的財務報表資訊，因此基本分析中對財務報表資訊的分析又為大宗，投資人將依據這些財務報表資訊決定未來投資方向，因此正確的「解讀財務報表資訊」為基本分析重要的課題。若近一步將「解讀財務報表資訊」此一環節分解，可以發現此環節之主要步驟為：
1. 選取財務報表中投資者認為重要的相關比率 
2. 依據過去經驗設定財務報表中相關比率的門檻條件
3. 符合條件者進行投資，可以發現上述步驟相當類似於「分類」的動作，投資人期望藉由財報資訊對眾多的標的物進行分類，而將結果分類為做多、做空、不操作等類別，而這些類別也反映了投資人的預期—未來會漲或是未來會跌，因此「如何分類」則是檢視財務報表資訊的核心動作。

> 因此以下將採取機器學習中分類演算法—隨機森林，應用於基本分析中對財務報表資訊的解讀，以財務報表資訊為自變數，股票報酬績效為應變數，建立分類模型，期望透過此分類模型，找出未來股價可能會漲之投資標的。

* 模型建構方式：

> 分類模型的建構則以季為一個時間單位，第t期的季報財務指標與t+1期相對應的預測標籤為一組，其中預測標籤為t期至t+1期間之股票報酬，使用資料之時間區間為2010年至2017年第二季結束，上述為使用於建構模型之所有樣本，考量公司營運過程可能遇到之季節性週期影響，因此模型訓練將使用特定季別之資料(進行此專題時處於2017年第三季財報公布後，因此以下使用第三季之資料進行模型訓練)。考量不同產業別之間可能有不同特性，模型將依據不同產業分別建立，且考量財報資料齊全程度與公司規模等問題，樣本資料只選擇上市公司，其中模型所需之樣本數量不宜過少，加上模型訓練將使用特定季別之資料(同產業內之樣本數將除以四)，因此必須選擇公司家數最多之產業別，下列為TEJ資料庫中前五大產業之上市公司家數：

> 建設：43家，半導體-IC設計：31家，汽車組件：26家，電子零組件-連接器：21家，資訊通路-零組件代理：18家

* 使用資料：

> 使用資料由TEJ獲得(資料整理過程未放入)，主要方式為在季報(單季)欄位之後加上本季財報公布至下季財報公布此期間內各股票之報酬率(做為要預測的標籤)

* 以下正式開始：

```python
#匯入資料及套件
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("data_set.csv")
data.head()
```

```python
#產業類別清單 → 數量前五大
data.TEJ子產業_名稱.unique()
```

```python
#以下以半導體IC設計業為例

data = data[data.TEJ子產業_名稱 == 'IC設計']
data.head()
```

```python
#將沒有要納入模型建構之X去除
data.drop(["公司", "TEJ子產業_名稱","date", "code"], axis = 1, inplace = True)
```

```python
#檢視IC設計業資料集中return(標籤)之分布狀況
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\Fonts\SimSun.ttc", size=15)

data["return"].hist(bins = 50, figsize = (10, 6), alpha = 0.7)
plt.xlabel(u"股價報酬", fontproperties=font)
plt.ylabel(u"數量", fontproperties=font)
plt.title('股票報酬直方圖', fontproperties=font, size = 20);
```

```python
#將原先之return標籤另存一份並分類，以下只分為兩類，報酬小於0.1者為第一類，報酬大於等於0.1者為第二類
Return = np.array(data["return"])

for i in range(0, len(Return)):
    if Return[i] < 0.1:
        Return[i] = 1
    else:
        Return[i] =2
        
#看各群組數量，兩組之數量不宜差距過多  
pd.DataFrame(Return).hist(figsize = (10, 6), alpha = 0.7)
plt.xlabel(u"類別", fontproperties=font)
plt.ylabel(u"樣本數", fontproperties=font)
plt.title('類別數量圖', fontproperties=font, size = 20)
```

```python
##以下準備資料集

#將return label拿掉
X = data.drop("return", axis = 1)

#將data轉成array
xx = np.zeros([len(data), 1])
for ix in X.columns:
    xx = np.append(xx, np.array(X["{}".format(ix)]).reshape(len(data), 1), axis = 1)
xx = xx[:, 1:]
X = xx

#將上述分好類之Return指定為Y(解答)
Y = Return
```

```python
#將 X 依據0.7, 0.3之比例分為訓練資料集及測試資料集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(X, Y, test_size = 0.3, random_state = 666)

#建立隨機森林模型
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion = "gini", 
                             n_estimators = 3,  
                             random_state = 20, 
                             max_features = 0.7, 
                             max_depth = 7,      
                             min_samples_leaf = 5) 

clf.fit(x_train, y_train)

#訓練資料集預測出之結果
y_predict_train = clf.predict(x_train)

#測試資料集預測出之結果
y_predict_test = clf.predict(x_test)

#評估訓練資料集與測試資料集之預測準確率
from sklearn.metrics import accuracy_score
train_acc = accuracy_score(y_train, y_predict_train )
test_acc = accuracy_score(y_test, y_predict_test)

print("Accuracy_train : %.2f%%" % (train_acc * 100.0),"\n"
      "Accuracy_test  : %.2f%%" % (test_acc * 100.0))
```

感覺準確率很高！也沒overfitting，要賺大錢了！但事實並不是這樣.......(最後會說明)

```python
#檢視混淆矩陣

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_predict_test)
```

```python
#視覺化混淆矩陣 - 網路找到的function
#網址：http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py

import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes) #, rotation=45
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plot_confusion_matrix(confusion_matrix(y_test, y_predict_test), classes = ["1", "2"], normalize = True)
```

在此模型下屬於類別二之資料預測能力稍微好一些。至於右上方之區塊則需要特別注意，由於模型預測為類別二，但實際卻為類別一，因此可能會出現實際損失；左下區塊則為潛在損失，由於模型預測為類別一，因此可能不納入考量，但其實際卻為類別二。

* 問題檢討：

> 1. 股價標籤之區間設定或許可以更早一些，股價應該會在財報公布之前已提前反應，因此可以將股價區間之設定向前移半個月或一個月

> 2. 資料問題：首先，在TEJ獲取的季報資料中，主要選擇獲利能力、償債能力與其他相關比率的部分，沒有一一篩選與考量財務指標間之相關性(對於模型建構應該會有影響)，其中若有缺值者則採取直接去除那列資料之方式，通常應該以補值之方式將資料補齊，但財報比率若用補值之方式感覺有些奇怪；接著財報中之數值有些為百分比有些則為非百分比之數值，對於此部分沒有特別處理，但或許對於模型之建構會有一定程度之影響

> 3. 上述模型建構時皆採取給定random state的數值下(分訓練資料集、測試資料集與模型建構時皆是)，嘗試調整各參數以最佳化模型，雖然此模型不論是訓練資料或測試資料之準確度皆接近82%，看起來很高且沒有overfitting的問題，但應該要透過Cross Validation的方式進行整體模型之評估(目前此部分之相關概念尚不熟，因此沒有如此操作)，否則雖然在此單一模型下表現很好，但可能有最佳化卻無法適用於未知資料的問題，無法達到機器學習Generalization之實際目標。下圖舉例，若將模型建構中之random state參數以真正隨機之方式進行，則可以發現準確率差異甚大：

```python
acc_train = []
acc_test = []

for i in (list(range(1,1001))):
    clf = RandomForestClassifier(criterion = "gini", 
                             n_estimators = 3,  
                             random_state = i, 
                             max_features = 0.7, 
                             max_depth = 7,      
                             min_samples_leaf = 5) 

    clf.fit(x_train, y_train)

    y_predict_train = clf.predict(x_train)

    y_predict_test = clf.predict(x_test)

    train_acc = accuracy_score(y_train, y_predict_train)
    test_acc = accuracy_score(y_test, y_predict_test)
    
    acc_train = np.append(acc_train, train_acc)
    
    acc_test = np.append(acc_test, test_acc)

#分別畫出在不同random state之下訓練資料集與測試資料集之準確率
plt.figure(figsize = (16, 10))
plt.plot(acc_train, label = "train_acc")
plt.plot(acc_test, label = "test_acc")
plt.legend()
```

由上圖可以發現在原先的模型下，高準確率只是剛好抓到在給定參數下，能夠最佳化準確率，且同時看似沒有overfitting的問題，但若改變隨機值則現出原形了...不論是overfitting或是預測資料集之極低準確率等問題都將現形。可能是資料處理過程導致garbage in garbage out的狀況、可能是模型參數設定的問題、也可能是財報資料本身便不適合拿來預測股價(歷史不一定會重演)。

```python
#從TEJ資料庫獲取股價資料
stock_price = pd.read_excel("Stockprice.xlsx", index_col = 1)

stock_price.head()
```

```python
#計算IC設計產業在2017年第三季財報公布至第四季財報公布間之股價報酬，計算方式2018/3/31股價除以2017/11/15股價後減1
stock_performance = stock_price.groupby("證期會代碼").apply(lambda x: (x["收盤價(元)"][len(x) - 1] / x["收盤價(元)"][0]) - 1).reset_index()
stock_performance.columns.values[1] = "Return"
```

這樣的結果其實在預期之內，由上述模型建構過程即可發現，在給定隨機值下才出現較高的準確率，而這樣的模型很有問題，應該要透過交叉檢驗的方式重新檢視模型，而這部分日後有時間將繼續研究，同時也須考量資料正規化的問題。以下為此次預測結果的混淆矩陣：

```python
#準備資料建立混淆矩陣
unrecommand_list_ic_desgin["Predict"] = 1 
unrecommand_list_ic_desgin["Real"] = unrecommand_list_ic_desgin.Return.apply(lambda x: 2 if x >=0.1 else 1)

recommand_list_ic_desgin["Predict"] = 2 
recommand_list_ic_desgin["Real"] = recommand_list_ic_desgin.Return.apply(lambda x: 2 if x >=0.1 else 1)

result = pd.concat([recommand_list_ic_desgin[["Predict", "Real"]],unrecommand_list_ic_desgin[["Predict", "Real"]]], axis = 0)

#計算混淆矩陣
confusion_matrix(result.Real, result.Predict)
```

### 以下重新進行分析，在建造模型時透過GridSearchCV尋找最佳模型參數，同時透過cross-validation進行檢驗

```python
#重新輸入資料(與上面步驟相同)

data = pd.read_csv("data_set.csv")
data = data[data.TEJ子產業_名稱 == 'IC設計']
data.drop(["公司", "TEJ子產業_名稱","date", "code"], axis = 1, inplace = True)

Return = np.array(data["return"])

for i in range(0, len(Return)):
    if Return[i] < 0.1:
        Return[i] = 1
    else:
        Return[i] =2
        
X = data.drop("return", axis = 1)
xx = np.zeros([len(data), 1])
for ix in X.columns:
    xx = np.append(xx, np.array(X["{}".format(ix)]).reshape(len(data), 1), axis = 1)
xx = xx[:, 1:]
X = xx
Y = Return

x_train, x_test, y_train, y_test =  train_test_split(X, Y, test_size = 0.3, random_state = 666)
```

```python
#建造模型，此部分採取新的做法
#使用GridSearchCV在給定所有的模型參數中，找尋最佳準確率的參數(類似於暴力解)
from sklearn.model_selection import GridSearchCV

#設定模型參數區間
clf_param_grid = {"n_estimators" : np.arange(10, 21, 1),
                  "criterion" : ["gini", "entropy"],
                  "max_features" : ["auto", "sqrt", "log2"],
                  "max_depth" : np.arange(1, 11, 1),
                  "min_samples_leaf" : np.arange(1, 6, 1)
                 }

#同樣選擇隨機森林演算法
clf = RandomForestClassifier(random_state = 666)

#將模型導入GridSearchCV
clf_GridSearchCV = GridSearchCV(param_grid = clf_param_grid,
                                estimator = clf,
                                scoring = "accuracy",
                                cv = 10,
                                n_jobs = 3)

#訓練模型
clf_GridSearchCV.fit(x_train, y_train)

#回傳準確率最高的模型參數及交叉驗證後的準確率
print("Best parameters found: ", clf_GridSearchCV.best_params_)
print("Best accuracy found: ", clf_GridSearchCV.best_score_)
```

```python
#選擇由上述過程找到的最佳模型
clf = clf_GridSearchCV.best_estimator_
# clf = RandomForestClassifier(criterion ='entropy', max_depth =  4, max_features = 'auto', min_samples_leaf = 1, n_estimators = 14, random_state = 666)


#訓練模型
clf.fit(x_train, y_train)

#訓練資料集預測出之結果
y_predict_train = clf.predict(x_train)

#測試資料集預測出之結果
y_predict_test = clf.predict(x_test)

#評估訓練資料集與測試資料集之預測準確率
from sklearn.metrics import accuracy_score
train_acc = accuracy_score(y_train, y_predict_train )
test_acc = accuracy_score(y_test, y_predict_test)

print("Accuracy_train : %.2f%%" % (train_acc * 100.0),"\n"
      "Accuracy_test  : %.2f%%" % (test_acc * 100.0))
```

訓練資料集93.02%的準確率未經交叉驗證，經由交叉驗證後的準確率為67.44%，與測試資料集60.71%的準確率相比，確實可能尚存在over fitting的問題

```python
#檢視混淆矩陣
confusion_matrix(y_test, y_predict_test)
```
